```python
# Task 3: 7 points
# Now we will build a linear model using scikit-learn's LinearRegression
# https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html

# 1 point for saving an instance of the LinearRegression class into linReg
# 3 points for the 'fit' method of the model
# 3 points for using the 'score' method of the model 

# If we don't do this, LinearRegression will complain
incRateReshaped = np.array(incRate).reshape(-1, 1)  

# First, we create and save an instance of the LinearRegression class
# See the web or the URL for LinearRegression above for an example.

linReg = LinearRegression()    # FINISH THIS LINE OF CODE

# Next, we call the the 'fit' method of that instance, and pass it
# the reshaped predictor variable and then the target variable for 
# the death rate. You already created and saved both of those values above.
# So you just need to find the correct variable names to use as the
# arguments to the 'fit' method.

linReg.fit(incRateReshaped, dthRate)



# In scikit-learn, we use the 'score' method of a model
# to find out its accuracy by passing it two arguments:
# the predictor data and the target variable, 
# i.e. the same two variables that you already passed to the 'fit' method.

linRegScore = linReg.score(incRateReshaped, dthRate) # FINISH THIS LINE OF CODE

print(f"The R2 value of a simple linear regression model is: {linRegScore}") # R2 SHOULD BE ABOUT 0.2019  Best possible is 1.0

# The next two lines show you how to extract the slope and the y intercept
# values from a fitted linear regression model with one predictor.
m = linReg.coef_[0]
b = linReg.intercept_

print(f"The slope of the fitted model is: {m}")
print(f"The y-intercept of the model is: {b}")
```

```python
# Now we will define several functions to manually compute the slope
# and y--intercept of the same model and below  
# you will compare them to what LinearRegression() just computed.

# I will give you the first function myself, and it's one of  
# the most important of all functions in data science: dot product

# The dot product of two input vectors x and y is a linear algebra 
# operation that computes the sum of their element-wise products.

# A vector is a sequence of numbers, e.g. an np.array in 1 dimension. 
# You will see examples of using vectors
# in the code that follows right AFTER this definition.
# Pay attention to this function and REMEMBER it because we will see
# the dot product operation in future weeks.  It's simple, but also
# very important and surprisingly VERSATILE, as you will see during this course.

# You should always include comments like the ones immediately under
# the first line of the function definition to explain what
# the function does, and what its argument variables are.

def dotProduct(x, y):
    # Compute the dot product of two equal-length vectors x and y
    # x and y are both either a single number or an np.array of numbers
    return(np.sum(x*y))
```

```python
# Task 8: 10 points

def slope(x, y, n): 
    # Compute the slope for a simple linear regression model where
    # x is a vector predictor, y is the vector response, and n is the number 
    # of observations, i.e. the number of rows in your data frame.
    # See instructions handout for math formula.  
    
    # As you think of how to write the code for the numerator and denominator
    # below, remember to USE THE FUNCTIONS DEFINED ABOVE. Each of THOSE functions can 
    # be used to help you define 'slope' here and/or yIntercept below.
    # So when you look at the math equations in the instructions handout for 
    # this assignment, think about how to relate the Python functions to the
    # math notation, and then USE those Python functions in the next two lines:
    
    numerator   = n * dotProduct(x, y) - np.sum(x) * np.sum(y)
    denominator =  n * sumSquares(x) - squareSum(x)
    
    return(numerator/denominator)
```

```python
# Task 9: 10 points   

def yIntercept(x, y, z):
    # Compute the value of the y intercept for a simple linear regression model where
    # x is the vector predictor, y is the vector response, and the third
    # argument z is the value computed by the function 'slope' above.
    
    return np.mean(y) - z * np.mean(x)  # See instructions handout for math formula
    
    
```
